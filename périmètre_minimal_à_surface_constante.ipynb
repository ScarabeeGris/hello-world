# Exercice : périmètre minimal à surface constante
$xy = 300$
 
$y  = \frac{300}{x}$
 
$l(x)= 2x + y = 2x + \frac{300}{x}$
 
$l(x)= \frac{2x^2 + 300}{x}$
 
$l'(x)= \frac {4x^2-(2x^2+300)}{x^2}=\frac{2x^2-300}{x^2}$
 
$2x^2-300 > 0$
 
$x^2 > 150$
 
$x > \sqrt{150}$

$l(x)$ est décroissante sur $]0;150[$

```
x = sqrt(150)
y = 300/x
print("La longueur de la cloture minimale (x = {:.2f} m et y = {:.2f} m) = {:.2f} m".format(x, y, 2*x+y))
print("Aire du champ : {:.2f} m2".format(x*y))
repère = plt.axes(xlim=(0, 50), ylim=(0, 50))
x = np.linspace(0.1, repère.get_xlim()[1],1000)
y = 300/x
xs = 12.25
repère.plot(x, y, label='xy=300')
repère.plot(x, -2*x+48.99, label='2x+y=48.99')
repère.plot(x, -2*x+30, label='2x+y=30')

repère.plot(xs, 300/xs, 'bo', label='Minimum')
plt.text(xs+1, (300/xs)+1, "P")
plt.plot([xs, xs],[0, 300/xs], color='k', linestyle=":")
plt.plot([0, xs],[300/xs, 300/xs], color='k', linestyle=":")

repère.set_xlabel('x (m)')
repère.set_ylabel('y (m)')
repère.set_title('Périmètre minimal pour une surface de 300')
repère.legend(loc="best", shadow=True)
plt.show()
```
